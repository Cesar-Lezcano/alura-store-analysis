### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



##Ingreso Total por cada Tienda:


```python
print(tienda.columns)
```

```python
ingreso1 = tienda["Precio"].sum()
ingreso2 = tienda2["Precio"].sum()
ingreso3 = tienda3["Precio"].sum()
ingreso4 = tienda4["Precio"].sum()

print(ingreso1)
print(ingreso2)
print(ingreso3)
print(ingreso4)
```

# 2. Ventas por categoría

```python
ventas_categoria1 = tienda["Categoría del Producto"].value_counts()
ventas_categoria2 = tienda2["Categoría del Producto"].value_counts()
ventas_categoria3 = tienda3["Categoría del Producto"].value_counts()
ventas_categoria4 = tienda4["Categoría del Producto"].value_counts()

print(ventas_categoria1)
print(ventas_categoria2)
print(ventas_categoria3)
print(ventas_categoria4)
```

# 3. Calificación promedio de la tienda


```python
promedio_valoracion1 = tienda["Calificación"].mean()
promedio_valoracion2 = tienda2["Calificación"].mean()
promedio_valoracion3 = tienda3["Calificación"].mean()
promedio_valoracion4 = tienda4["Calificación"].mean()

print(promedio_valoracion1)
print(promedio_valoracion2)
print(promedio_valoracion3)
print(promedio_valoracion4)
```

# 4. Productos más y menos vendidos

```python
# Productos más vendidos (top 10)
mas_vendidos1 = tienda["Producto"].value_counts().head(10)
mas_vendidos2 = tienda2["Producto"].value_counts().head(10)
mas_vendidos3 = tienda3["Producto"].value_counts().head(10)
mas_vendidos4 = tienda4["Producto"].value_counts().head(10)

print("Tienda 1 - Más vendidos")
print(mas_vendidos1)

print("\nTienda 2 - Más vendidos")
print(mas_vendidos2)

print("\nTienda 3 - Más vendidos")
print(mas_vendidos3)

print("\nTienda 4 - Más vendidos")
print(mas_vendidos4)

# Productos menos vendidos (bottom 10)
menos_vendidos1 = tienda["Producto"].value_counts().tail(10)
menos_vendidos2 = tienda2["Producto"].value_counts().tail(10)
menos_vendidos3 = tienda3["Producto"].value_counts().tail(10)
menos_vendidos4 = tienda4["Producto"].value_counts().tail(10)

print("\nTienda 1 - Menos vendidos")
print(menos_vendidos1)

print("\nTienda 2 - Menos vendidos")
print(menos_vendidos2)

print("\nTienda 3 - Menos vendidos")
print(menos_vendidos3)

print("\nTienda 4 - Menos vendidos")
print(menos_vendidos4)
```

# 5. Envío promedio por tienda

```python
envio_promedio1 = tienda["Costo de envío"].mean()
envio_promedio2 = tienda2["Costo de envío"].mean()
envio_promedio3 = tienda3["Costo de envío"].mean()
envio_promedio4 = tienda4["Costo de envío"].mean()

print(envio_promedio1)
print(envio_promedio2)
print(envio_promedio3)
print(envio_promedio4)
```

## Visualización

```python
import matplotlib.pyplot as plt

# Ingresos por tienda
ingresos = [
    tienda["Precio"].sum(),
    tienda2["Precio"].sum(),
    tienda3["Precio"].sum(),
    tienda4["Precio"].sum()
]

plt.figure()
plt.bar(["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"], ingresos)
plt.title("Ingresos por tienda")
plt.xlabel("Tienda")
plt.ylabel("Ingreso total")
plt.show()

# Distribución de categorías (tienda 1)
categorias = tienda["Categoría del Producto"].value_counts()

plt.figure()
plt.pie(categorias, labels=categorias.index, autopct='%1.1f%%')
plt.title("Distribución de categorías (Tienda 1)")
plt.show()

# Calificación promedio por tienda
promedios = [
    tienda["Calificación"].mean(),
    tienda2["Calificación"].mean(),
    tienda3["Calificación"].mean(),
    tienda4["Calificación"].mean()
]

plt.figure()
plt.bar(["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"], promedios)
plt.title("Calificación promedio por tienda")
plt.xlabel("Tienda")
plt.ylabel("Calificación promedio")
plt.show()
```
